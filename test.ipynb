```python
model1 = torch.load('best_49.pt')
model2 = torch.load('best_50.pt')
model3 = torch.load('best_51.pt')
```

```python
import copy
model = copy.deepcopy(model1)

model['model'].state_dict()
```

```python
models = [model1['model'], model2['model'], model3['model']]
```

```python
fed_dict = {}
for key in model['model'].state_dict().keys():
    fed_dict[key] = sum([model1['model'].state_dict()[key], model2['model'].state_dict()[key], model3['model'].state_dict()[key]])/3

model['model'].load_state_dict(fed_dict)
model['model'].state_dict()
```
